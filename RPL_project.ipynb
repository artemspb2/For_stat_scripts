```python
# Создадим колонку, в которой будут сохранены значения времени проведения матчей в качестве категориальной пременной
# Дневные - для матчей до 17.00, вечерние - для матчей в 17.00 и позже 
# Объявим функцию
def get_period(time):
    if time >= pd.to_datetime("17:00:00", format="%H:%M:%S").time():
        return "вечерний"
    else:
        return "дневной"
```

```python
# Теперь можно создать столбец с результатом матча (победа хозяев, гостей или ничья)
import numpy as np

conditions = [
    (rpl['home_team_goals'] > rpl['away_team_goals']),
    (rpl['home_team_goals'] < rpl['away_team_goals']),
    (rpl['home_team_goals'] == rpl['away_team_goals'])
]

values = ['home_win', 'away_win', 'tie']

rpl['result'] = np.select(conditions, values)
```

```python
# Но для начала надо определить силу команд. Как это сделать?
# Мы уже создали таблицу с командами и количеством забитых голов  в домашних матчах. Там же можно создать разницу забитых и пропущенных мячей 
# и на ее основе проранжировать команды по силе, создать столбец с количественной, ранговой или категориальной переменной, показывающей среднюю силу команды в сезоне. 
# Потом объединить рассчитанную таблицу с нашей основной таблицей rpl.  Но это для совсем ленивых.
# Мы сделаем немного по-другому: сдлаем новую таблицу с количеством очков в конце сезона и это буде показателем силы команды.
```

```python
# Создадим временный датасет с результатами матчей, 3 столбцами, рассчитанными по столбцу result. 
# 1 - если это значение справедливо для столбца result, 0 - если нет
temp_df = pd.get_dummies(rpl['result'])
```

```python
# Создадим временную таблицу с результатми домашних матчей для каждой команды РПЛ
t1 = rpl.groupby('home_team', as_index = False).agg({'home_win':'sum', 'tie':'sum', 'away_win':'sum'})
t1
```

```python
# Создадим временную таблицу с результатми выездных матчей для каждой команды РПЛ
t2 = rpl.groupby('away_team', as_index = False).agg({'home_win':'sum', 'tie':'sum', 'away_win':'sum'})
t2
```

```python
# В каждой таблице создадис столбец с рассчитанным количеством набранных очков
t1['points'] = t1['home_win']*3 + t1['tie']*1
t2['points'] = t2['away_win']*3 + t2['tie']*1
```

```python
# Проверяю, что все получилось
t1.head(3)
```

```python
t2.head(3)
```

```python
# Теперь можно объединить таблицы
```

```python
# У нас одинаковые названия столбцов, надо переименовать перед объединением
t1 = t1.rename(columns = {'home_win' : 'h_home_win', 'tie' : 'h_tie', 'away_win' : 'h_away_win', 'points' : 'h_points'})
t2 = t2.rename(columns = {'home_win' : 'a_home_win', 'tie' : 'a_tie', 'away_win' : 'a_away_win', 'points' : 'a_points'})
```

```python
# Объединим,удалим лишнюю колонку с названием, а еще одну колонку с названием переименуем, 
# тк в нашей таблице будут подсчитаны очки и домашних и выездных матчей
final_table = pd.concat([t1, t2], axis=1).drop('away_team', axis = 1).rename(columns = {'home_team' : 'team'})
final_table
```

```python
# Рассчитаем количество побед, ничей, поражений для всего сезона
final_table['final_wins'] = final_table.h_home_win + final_table.a_home_win
final_table['final_ties'] = final_table.h_tie + final_table.a_tie
final_table['final_losses'] = final_table.h_away_win + final_table.a_away_win

final_table['final_points'] = final_table.h_points + final_table.a_points
```

```python
# Сделаем сортировку по количеству набранных очков, сравним с данными: https://www.championat.com/football/_russiapl/tournament/4465/
# Сходится
final_table.sort_values('final_points', ascending = False)
```

```python
# У нас есть красивая табличка с победами, ничьими, поражениями дома, на выезде и в целом в сезоне. 
# Хотя полностью она нам не нужна, но и отдельно представлетя собой интерес и результат.
```

```python
# Чтобы не запутаться переименую названия столбцов там, где неочевидно что к чему
# Теперь h_win	h_tie	h_loss	h_points относятся к домашним результатам команды, представленной в строке
# А a_loss	a_tie	a_win	a_points к выездным

final_table = final_table.rename(columns = {'h_home_win': 'h_win',  'h_away_win': 'h_loss', 'a_home_win': 'a_loss', 'a_away_win': 'a_win'}).sort_values('final_points', ascending = False)
final_table
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf 


# Я изучал R и мне нравится этот язык программирования для статистики, этот синтаксис напоминает модели в R, поэтому выбрал его
# Здесь я не буду объяснять что такое логистическая регрессия - можно посмотреть в сети Интернет
# Напомню, что в процессе разведывательного анализа данных мне интересно было - есть ли связь между победами домашней команды и 
# временем года, в которое проходит матч. Также я предположил, что, возможно, просто на зиму выпало больше игр слабых команд 
# с сильными, которые играли на выезде. Я не стал смотреть календарь (хотя мог бы), а решил использовать логистическую регресию,
# с помощью который мог бы подтвердить или опровергнуть свою гипотезу статистически.
# Поэтому в предикторах у меня 2 фактора: 1.отношение силы команды-хозяйки и команды-гостя, 2 - время года, в которое сыгран матч.
# Вторая переменная категориальная, для нее в саммари лог. регрессии выбирается одно из значений как опорное, относительно которого 
# считаются статистики других значений переменной.
results = smf.logit('home_win ~ home_to_away_power_std + season', rpl).fit()
print(results.summary())
```

```python
# Мы объяснили 16% дисперсии (Pseudo R-squ.: 0.1634), в социальных науках это неплохой результат
# Фактор соотношения сил команд значим, сезоны - нет.
```

```python
# Интересно посмотреть сколько дисперсии объяснит модель без переменной с временами года
results2 = smf.logit('home_win ~ home_to_away_power_std', rpl).fit()
print(results2.summary())
# 14%
```

```python
# Также интересно - есть ли интерактивные эффекты, взаимодействие пременных?
results3 = smf.logit('home_win ~ home_to_away_power_std * season', rpl).fit()
print(results3.summary())
```

```python
# Хотя (Pseudo R-squ.:  0.1772) стал больше, но мы на это потратили слишком много степеней свободы 
# (есть тесты, доказывающие это статистически, но здесь это и так видно). Игра не стоит свеч. 
# Статзначимых интерактивных эффектов также не нашли
# Лучше всего результат матча предсказывает сила команд (из 2 исследованных переменных), а сезонность не играет роли.
```

```python
# Ну и мы видим, что и сама переменная сезонность даже без наличия мощного предиктора-конкурента не может объяснить 
# серьезной доли дисперсии зависимой переменной.  Pseudo R-squ.: 0.005009. 
# Так что особого смысла ставить этот предиктор в модель и не было.
# Гораздо интереснее бывают ситуации, когда по отдельности предикторы хорошо объясняют зависимую переменную,
# но при добавлении в модель один из них станосится незначим
results4 = smf.logit('home_win ~ season', rpl).fit()
print(results4.summary())
```
